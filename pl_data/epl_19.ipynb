```python
# read the epl19.csv

import pandas as pd
import numpy as np
df = pd.read_csv('epl19.csv', encoding = 'utf8')
```

```python
# create a mini dataFrame object to drop betting odds columns

dfMini = df[["Date", "Time", "HomeTeam", "AwayTeam", "FTHG", "FTAG","FTR", "Referee", "HS", "HST", "AS", "AST"]]
```

```python
# create a pivot table to show the total number of home goals scored by each team sorted in descending order

homeGoals = dfMini.pivot_table(values = 'FTHG', index = 'HomeTeam', aggfunc = 'sum').sort_values(by = 'FTHG', ascending = False)
```

```python
# import the matplot library
import matplotlib.pyplot as plt
```

the .bar() method plots the table data. Two arguments -
 
1. x - the x coordinate for each bar. we provide a sequence from 1-20 for a 20 bar chart. the np.arrange provides evenly spaced values within a given interval.

2. height - the value of each bar, i.e. the goals column of the table

```python
# displays homeGoals raw data as a bar chart

plt.bar(x=np.arange(1,21),height=homeGoals['FTHG'])
```

```python
teamColours = ['#C8102E', '#6CABDD', '#003090', '#DA291C', '#132257', '#EF0107', '#7A263A', '#034693', '#FDB913', '#6C1D45', '#003399', '#FFF200', '#95BFE5', '#0057B8', '#EE2737', '#DA291C',  '#D71920', '#FBEE23', '#1B458F', '#241F20']

plt.bar(x=np.arange(1,21),height=homeGoals['FTHG'], color = teamColours)
plt.title("Premier League 18/19 Home Goals")
plt.xticks(np.arange(1,21), homeGoals.index, rotation = 90)
plt.xlabel("Home Team")
plt.ylabel("Goals")
plt.show()
```
